```python
import markovify

text = """
Data science is an interdisciplinary field that uses scientific methods.
It involves statistics, data analysis, machine learning and artificial intelligence.
Data science helps in extracting meaningful insights from structured and unstructured data.
Machine learning is a subset of artificial intelligence that focuses on building models.
Artificial intelligence is transforming industries across the world.
Big data technologies are used to process large datasets efficiently.
Data analysts and scientists work together to solve complex problems.
"""

# Build model (slightly relaxed)
model = markovify.Text(text, state_size=1)

print("Generated Sentences:\n")

for i in range(5):
    sentence = model.make_short_sentence(100)
    if sentence:
        print(sentence)
    else:
        print("Could not generate sentence")
```
